```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
```

```python
sentence=["I","Love","Deep","Learning"]
print(sentence)
word_embeddings={
    "I":1.0,
    "Love":2.0,
    "Deep":3.0,
    "Learning":4.0
}
print(word_embeddings)
x=np.array([
    [
    [word_embeddings["I"]],
    [word_embeddings["Love"]],
    [word_embeddings["Deep"]],
    [word_embeddings["Learning"]]
    ]
])
print(x)
```

```python
y=np.array([word_embeddings["Learning"]])
print(y)
print("Input Shape",x.shape)
print("output shape",y.shape)
```

```python
#Build SimpleRNN Model
model=Sequential()
model.add(SimpleRNN
            (units=8,
            activation='tanh',
             input_shape=(3,1)))
model.add(Dense(1))
```

```python
model.compile(
    optimizer='adam',
    loss='mse'
)
```

```python
model.fit(x,y,epochs=300,verbose=1)
```

```python
test_sequence=np.array([
    [
        [1.0],
        [2.0],
        [3.0]
    ]
])
prediction=model.predict(test_sequence)
print("prediction:",prediction[0][0])
```

```python
predicted_number=prediction[0][0]
closest_word=min(word_embeddings,key=lambda word:abs(word_embeddings[word]-predicted_number))
print("The predicted word is:",closest_word)
```
